# Data introduction

https://data.sfgov.org/Public-Safety/Police-Department-Incident-Reports-Historical-2003/tmnf-yvry/about_data

This dataset is a collection of crime activity in San Francisco...

The data is limited to the time period 2003 to 2018 ...

```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv('data/sf_data.csv')
steal_crimes = ['STOLEN PROPERTY', 'ROBBERY', 'BURGLARY', 'VEHICLE THEFT', 'LARCENY/THEFT']
df = df.loc[df['Category'].isin(steal_crimes)]
df['Date'] = pd.to_datetime(df['Date'])
df = df[df['Date'].dt.year < 2018]
```

```python
grouped = df['Category'].groupby(df['Date'].dt.year).value_counts().unstack().fillna(0)
```

```python
grouped = grouped[steal_crimes]
```

# Visualization 1 - Overview (static time-series plot)

Let's get an overview of the distribution of different stealing-related crimes over the years

```python
colors = plt.cm.gist_rainbow(np.linspace(0,1,5))
```

```python
colors[0]
```

```python
sns.set_style('dark')

big_ax = plt.subplot2grid((4,3), (0, 0), rowspan=5, colspan=2)
big_ax.set_title('Big Plot')


grouped.plot(kind='bar', stacked=True, figsize=(10,7), colormap='gist_rainbow', ax=big_ax)
plt.title('Steal-related crimes in San Francisco', fontsize=16, pad=10)
big_ax.set_xlabel('Year', fontsize=12, labelpad=10)
# plt.xticks(rotation=45)
big_ax.set_ylabel('Number of crimes', fontsize=12, labelpad=10)
# plt.grid(axis='y')
# plt.legend(loc='upper center', ncols=5, bbox_to_anchor=(0.5, 1.07))
big_ax.legend(loc='lower center', ncols=5, bbox_to_anchor=(0.5, -0.3))


# Create two smaller plots stacked on the right
ax1 = plt.subplot2grid((5,4), (0, 3))
grouped[steal_crimes[0]].plot(kind='bar', ax=ax1, legend=False, color=colors[0])
ax1.set_yticks([])
ax1.set_xticks([])
ax1.set_xlabel('')

ax2 = plt.subplot2grid((5,4), (1, 3))
grouped[steal_crimes[1]].plot(kind='bar', ax=ax2, legend=False, color=colors[1])
ax2.set_yticks([])
ax2.set_xticks([])
ax2.set_xlabel('')

ax3 = plt.subplot2grid((5,4), (2, 3))
grouped[steal_crimes[2]].plot(kind='bar', ax=ax3, legend=False, color=colors[2])
ax3.set_yticks([])
ax3.set_xticks([])
ax3.set_xlabel('')

ax4 = plt.subplot2grid((5,4), (3, 3))
grouped[steal_crimes[3]].plot(kind='bar', ax=ax4, legend=False, color=colors[3])
ax4.set_yticks([])
ax4.set_xticks([])
ax4.set_xlabel('')

ax5 = plt.subplot2grid((5,4), (4, 3))
grouped[steal_crimes[4]].plot(kind='bar', ax=ax5, legend=False, color=colors[4])
ax5.set_yticks([])
# change xticks fontsize

ax5.set_xlabel('')


plt.tight_layout()
plt.show()
```
